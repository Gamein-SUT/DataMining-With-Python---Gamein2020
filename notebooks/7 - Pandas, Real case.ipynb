```python
df1_clean = pd.DataFrame()
```

```python
def movie_detail(mstr):
    try:
        mstr = eval(mstr)
        try:
            C = mstr['Country']
        except:
            C = np.nan

        try:
            L =mstr['Language']
        except:
            L = np.nan

        try:
            AKA = mstr['Also Known As']
        except:
            AKA = np.nan
            
        try:
            B = mstr['Budget']
        except:
            B = np.nan

        try:
            R = mstr['Runtime']
        except:
            R = np.nan
            
        return C, L, AKA, B, R
    except:
        return np.nan,np.nan,np.nan,np.nan,np.nan
```

```python
df1_clean['Country'] = df1['Movie detail'].apply(movie_detail).apply(lambda x:x[0])
df1_clean["Language"] = df1['Movie detail'].apply(movie_detail).apply(lambda x:x[1])
df1_clean["AKA"] = df1['Movie detail'].apply(movie_detail).apply(lambda x:x[2])
df1_clean["Buget"] = df1['Movie detail'].apply(movie_detail).apply(lambda x:x[3])
df1_clean['Runtime'] = df1['Movie detail'].apply(movie_detail).apply(lambda x:x[4])
```

```python
df1_clean.head()
```

```python
df1_clean['Country']
```

```python
def spliter_striper(mstr):
    try:
        mstr = mstr.split("|")
        res = []
        for i in mstr:
            res.append(i.strip())
        return ','.join(res)
    except:
        return np.nan
        
```

```python
df1_clean['Country'] = df1_clean['Country'].apply(spliter_striper)
```

```python
df1_clean['Language'] = df1_clean['Language'].apply(spliter_striper)
```

```python
df1_clean['Buget'][0]
```

```python
def Bunit(mstr):
    try:
        if mstr[0] == '$':
            return '$'
        else:
            return mstr[:3]
    except:
        return np.nan
```

```python
df1_clean['Buget-Unit'] = df1_clean['Buget'].apply(Bunit)
```

```python
def Bext(mstr):
    try:
        mstr = mstr.replace(',','')
        res = re.findall(r'(\d+)',mstr)
        return res[0]
    except:
        return np.nan
```

```python
df1_clean['Buget'] = df1_clean['Buget'].apply(Bext)
```

```python
df1_clean
```

```python
df1_clean['Runtime']
```

```python
def runtime_val(mstr):
    try:
        mstr = mstr.split("|")
        mstr = mstr[0]
        adad = re.findall(r'(\d+)',mstr)[0]
        return adad
    except:
        return np.nan
        
```

```python
def runtime_unit(mstr):
    try:
        mstr = mstr.split("|")
        mstr = mstr[0]
        vahed = re.findall(r'\d+\s*(\w+)','137 min')
        return vahed[0]
    except:
        return np.nan
```

```python
df1_clean['Runtime'] = df1_clean['Runtime'].apply(runtime_val)
```

```python
df1_clean['Runtime-vahed'] = df1_clean['Runtime'].apply(runtime_unit)
```
