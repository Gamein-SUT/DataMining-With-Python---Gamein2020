# Central Limit Theorem

```python
x = np.random.uniform(0,1,10000)
plt.plot(x,'b.')
```

```python
plt.hist(x,5);
```

```python
x = np.random.uniform(0,1,1000)
for _ in range(9000):
    x += np.random.uniform(0,1,1000)
x -= x.mean()
x /= x.std()
plt.hist(x,30);
```

# Maximum likelihood estimation

![image](https://wikimedia.org/api/rest_v1/media/math/render/svg/6af346398849ba9dea7df6a7e0f07da49f3c34b5)

```python
def add_noise(x,sigma):
    return 0.7*x+np.random.normal(0,sigma,x.shape)
```

```python
age = np.random.uniform(20,50,10000)
# fc = age
fc = add_noise(age,4)
```

```python
plt.plot(age,fc,'.')
```

```python
ss = fc-0.7*age
plt.hist(ss);
```

```python
age = np.random.uniform(20,50,10000)
```

```python
stamina = (50-age+20)/50
age = add_noise(age,0.0)
```

```python
stamina = add_noise(stamina,0.0)
```

```python
plt.plot(age,stamina,'.')
```

```python
age = np.random.uniform(20,50,100)
h1 = np.random.normal(300,100,100)
age.std(),h1.std()
```

```python
stamina = (50-age+20)/50 + 0.01*h1
```

```python
plt.plot(age,stamina,'.')
```

```python
plt.plot(h1,stamina,'.')
```

```python
x = np.random.normal(0,1,100)
y = np.sin(2*x)+np.random.normal(0,0.1,100)
```

```python
plt.plot(x+3,y,'.')
```

```python
coefs, cov = np.polyfit(x, y, deg=5, cov=True)
```

```python
plt.plot(x+3,y,'.')
plt.plot(x+3,np.polyval(coefs,x),'r.')
```
