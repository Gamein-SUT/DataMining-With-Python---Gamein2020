```python
male = ['Under_18_MALE_SCORE_DEMOGRAPHIC_GG_RATING',
 '18_29_MALE_SCORE_DEMOGRAPHIC_GG_RATING',
 '30_44_MALE_SCORE_DEMOGRAPHIC_GG_RATING',
 'Over_45_MALE_SCORE_DEMOGRAPHIC_GG_RATING']

female = ['Under_18_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING',
 '18_29_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING',
 '30_44_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING',
 'Over_45_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING']
```

```python
dff = pd.DataFrame()
rating_val = []
rating_cat = []
rating_sex = []
for i in male:
    rating_val += list(df[i].values)
    temp = i.replace('Under_18_MALE_SCORE_DEMOGRAPHIC_GG_RATING', '-18')
    temp = temp.replace('18_29_MALE_SCORE_DEMOGRAPHIC_GG_RATING', '18-29')
    temp = temp.replace('30_44_MALE_SCORE_DEMOGRAPHIC_GG_RATING', '30-44')
    temp = temp.replace('Over_45_MALE_SCORE_DEMOGRAPHIC_GG_RATING', '+45')
    rating_cat += [temp]*len(df[i])
    rating_sex += ['Male']*len(df[i])
    
for i in female:
    rating_val += list(df[i].values)
    temp = i.replace('Under_18_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING', '-18')
    temp = temp.replace('18_29_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING', '18-29')
    temp = temp.replace('30_44_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING', '30-44')
    temp = temp.replace('Over_45_FEMALE_SCORE_DEMOGRAPHIC_GG_RATING', '+45')
    rating_cat += [temp]*len(df[i])
    rating_sex += ['Female']*len(df[i])
```

```python
dff['Rating'] = rating_val
dff['AC'] = rating_cat
dff['Gender'] = rating_sex
```

```python
fig,ax = plt.subplots(figsize = (12,8))
ax = sns.violinplot(data = dff, x='Rating', y = 'AC', hue='Gender',split=True)
```
