# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sqrt(x)
```

```python
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
plt.show()
```

```python
plt.savefig("mplot2.pdf")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
plt.plot(x,y,'b--', label="sqrt")
plt.plot(y,x,'r*-', label="tavan 2")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
plt.legend(loc='best')
plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes.plot(x,y, 'b', label="b-")
axes.plot(y,x, 'r--', label="r--")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title")
axes.legend()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.4,0.2, 0.4, 0.3])

axes1.plot(x,y, 'b', label="b-")

axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("Title")
axes1.legend()

axes2.plot(y,x, 'r--', label="r--")



axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Title")
axes2.legend()
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,y)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(5,5))

axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')


plt.tight_layout()
```

```python
fig.savefig("myplot.pdf", dpi=300)
```

```python
canvas, ax = plt.subplots(figsize=(12,8))


ax.plot(x,x+1, color='red', linewidth=0.25)
ax.plot(x,x+2, color='red', lw=0.55)
ax.plot(x,x+3, color='red', linewidth=0.75)


ax.plot(x,x+4, color='green', linewidth=0.25, linestyle='-')
ax.plot(x,x+5, color='green', lw=0.55, ls = '-.')
ax.plot(x,x+6, color='green', linewidth=0.75, ls = ':')

ax.plot(x,x+7, color='green', linewidth=0.25, linestyle='-', marker = "+")
ax.plot(x,x+8, color='green', lw=10.55, ls = '-.', marker = "o")
ax.plot(x,x+9, color='green', linewidth=0.75, ls = ':', marker = "1", ms=20)
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap="coolwarm", lw = 10)
```

```python
tips['all'] = ""
sns.violinplot(x = "all", y='total_bill',data=tips,hue = 'sex', split=True)
```
